We have separeted the Target varaible and the independent variable in x and y and now we will standardize our independent variables through Standard Scaler.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x1 = pd.DataFrame(sc.fit_transform(x), columns = x.columns)
```

Now as we have standardize our data we will check for Variance Inflation Factor.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()

vif['Values'] = [variance_inflation_factor(x1.values,i) for i in range(len(x1.columns))]
vif['Features'] = x1.columns
```

The vif value seems to be very high for few columns, we will be dropping the column 'ER' which is having highest vif value.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y, test_size = 0.30, random_state = 42 )
```

```python
lr = LinearRegression()

lr.fit(x_train,y_train)
lrpred = lr.predict(x_test)
lrsc = r2_score(y_test, lrpred)
lrmse = mean_squared_error(y_test, lrpred)
lrmae = mean_absolute_error(y_test, lrpred)

print('The Score for Linear Regression is ', lrsc.round(2))
print('The Mean Sqaured Error for Linear Regression is ', lrmse)
print('The Mean Absolute Error for Linear Regression is', lrmae)
```

Importing Other Regression Models

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Lasso,Ridge
from sklearn.ensemble import ExtraTreesRegressor
```

```python
dtr = DecisionTreeRegressor()

dtr.fit(x_train,y_train)
dtrpred = dtr.predict(x_test)
dtrsc = r2_score(y_test, dtrpred)
dtrmse = mean_squared_error(y_test, dtrpred)
dtrmae = mean_absolute_error(y_test, dtrpred)

print(dtrsc.round(2))
print(dtrmse)
print(dtrmae)
```

```python
rfr = RandomForestRegressor()

rfr.fit(x_train,y_train)
rfrpred = rfr.predict(x_test)
rfrsc = r2_score(y_test, rfrpred)
rfrmse = mean_squared_error(y_test,rfrpred)
rfrmae = mean_absolute_error(y_test,rfrpred)

print(rfrsc.round(2))
print(rfrmse)
print(rfrmae)
```

```python
svr = SVR()

svr.fit(x_train,y_train)
svrpred = svr.predict(x_test)
svrsc = r2_score(y_test, svrpred)
svrmse = mean_squared_error(y_test, svrpred)
svrmae = mean_absolute_error(y_test, svrpred)

print(svrsc.round(2))
print(svrmse)
print(svrmae)
```

```python
gbr = GradientBoostingRegressor()

gbr.fit(x_train,y_train)
gbrpred = gbr.predict(x_test)
gbrsc = r2_score(y_test, gbrpred)
gbrmse = mean_squared_error(y_test,gbrpred)
gbrmae = mean_absolute_error(y_test, gbrpred)

gbrsc.round(2), gbrmse, gbrmae
```

```python
knr = KNeighborsRegressor()

knr.fit(x_train,y_train)
knrpred = knr.predict(x_test)
knrsc = r2_score(y_test, knrpred)
knrmse = mean_squared_error(y_test, knrpred)
knrmae = mean_absolute_error(y_test, knrpred)

knrsc.round(2), knrmse, knrmae
```

```python
ls = Lasso()

ls.fit(x_train, y_train)
lspred = ls.predict(x_test)
lssc = r2_score(y_test, lspred)
lsmse = mean_squared_error(y_test, lspred)
lsmae = mean_absolute_error(y_test, lspred)

lssc.round(2), lsmse, lsmae
```

```python
ri = Ridge()

ri.fit(x_train, y_train)
ripred = ri.predict(x_test)
risc = r2_score(y_test, ripred)
rimse = mean_squared_error(y_test, ripred)
rimae = mean_absolute_error(y_test, ripred)

risc.round(2), rimse, rimae
```

```python
etr = ExtraTreesRegressor()

etr.fit(x_test, y_test)
etrpred = etr.predict(x_test)
etrsc = r2_score(y_test, etrpred)
etrmse = mean_squared_error(y_test, etrpred)
etrmae = mean_absolute_error(y_test, etrpred)

etrsc.round(2), etrmse, etrmae
```

So the below are r2_score, Mean Squared Error and Mean Absolute Error.

```python
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Liner Regression is', lrsc.round(2), lrmse.round(2), lrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Decision Tree Regressor is', dtrsc.round(2), dtrmse.round(2), dtrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Random Forest Regressor is', rfrsc.round(2), rfrmse.round(2), rfrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Support Vector Regressor is', svrsc.round(2), svrmse.round(2), svrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Gradient Boosting Regressor is', gbrsc.round(2), gbrmse.round(2), gbrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for KNeighbors Regressor is', knrsc.round(2), knrmse.round(2), knrmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Lasso', lssc.round(2), lsmse.round(2), lsmae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Ridge', risc.round(2), rimse.round(2), rimae.round(2))
print('The Accuracy Score, Mean Squared Error & Mean Absolute Error for Extra Trees Regressor', etrsc.round(2), etrmse.round(2), etrmae.round(2))
```
